TEXT SUMMARIZATION

```python
print(text.count("."))
nopunct =[char for char in text if char not in string.punctuation]
nopunct = "".join(nopunct)
nopunct
```

```python
# removing the stopwords from the text
processed_text = [word for word in nopunct.split() if word.lower() not in nltk.corpus.stopwords.words("english")]
processed_text
```

```python
# ranking the words according to the word frequency
wordFreq = {}
for word in processed_text:
  if word not in wordFreq:
    wordFreq[word]=1
  else:
    wordFreq[word] +=1
wordFreq
```

word frequency

```python
print(dict(sorted(wordFreq.items(),key = lambda item:item[1], reverse = True)))
```

sentenct frequency

```python
sent_list = nltk.sent_tokenize(text)
sent_list
```

```python
sent_score = {}
for sent in sent_list:
  for word in nltk.word_tokenize(sent.lower()):
    if word in wordFreq.keys():
      if sent not in sent_score.keys():
        sent_score[sent] = wordFreq[word]
      else:
        sent_score[sent] = sent_score[sent] + wordFreq[word]
```

```python
sent_score
```

```python
print(dict(sorted(sent_score.items(),key = lambda item:item[1], reverse = True)))
```

```python
summary_sent = nlargest(3,sent_score, key = sent_score.get)
summary_sent = " ".join(summary_sent)
summary_sent
```
